# Fitting a nueral net

### Read the cleaned dataset

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
df = pd.read_csv('../clean_dataset.csv')
```

### Remove old index

```python
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

### Define x and y

```python
y = df.iloc[:,0:7]
x = df.iloc[:,7:]
x.head(10)
```

### Split testing and training data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

### Build neural net

```python
nn=MLPClassifier(activation='logistic', solver='sgd',hidden_layer_sizes=(100,150),random_state=1)
```

```python
nn.fit(x_train,y_train)
```

### Test neural net

```python
prediction=nn.predict(x_test)
prediction
```

```python
a = y_test.values
```

```python
count=0
```

```python
for x in range(prediction.shape[0]):
    for y in range(prediction.shape[1]):
        if prediction[x][y]==a[x][y]:
            count=count+1
```

```python
count
```

### Success Rate

```python
success_rate = count / 61551
success_rate
```
