# Fitting a neural net

### Read the cleaned dataset

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MultiLabelBinarizer
df = pd.read_csv('../clean_dataset.csv')
```

### Define x and y

```python
y = df.iloc[:,0:7]
x = df.iloc[:,7:]
x.head(10)
```

### Split testing and training data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

### Build neural net

```python
nn=MLPRegressor(activation='logistic', solver='sgd',hidden_layer_sizes=(1),random_state=1)
```

```python
nn.fit(x_train,y_train)
```

### Test neural net

```python
prediction = nn.predict(x_test)
y_test.shape
for i in range(len(prediction)):
    print('actual:\n', y_test.iloc[i])
    print('prediction: ', y_test.columns[np.argmax(prediction[i])])
    print()
```

```python
# I would avoid rolling your own scoring functions (accuracy, ROC, AROC, lift, etc.)
# MLPClassifier has a built in score function
# and scikit-learn has a bunch of score functions built in
# https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html
nn.score(x_test, y_test)
```
